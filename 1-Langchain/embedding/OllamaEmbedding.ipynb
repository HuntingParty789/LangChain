Ollama supports embedding models, making it possible to build retieval augmented generation (RAG) applications that combines txt prompts with existing documents or other data. 

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma:2b") ## by default it uses lama 2
)
```

```python
r1=embeddings.embed_documents(
    [
        "Goku is the main protagonist of the Dragon Ball series.",
        "Vegeta is the prince of all Saiyans and a rival to Goku.",
    ]
)
```

```python
len(r1[0])
```
