### Sentene Transformers on HuggingFace

HuggingFace sentence-transformers is a python framework for state-of-the-rt sentence, text and image embeddings. One of the embedding models is used in the HuggingFaceEmbeddings class. We have also an alias for SetenceTransformerEmbeddings for users who are more familiar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text="This is a test sentence for embedding."
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)  # Should return the length of the embedding vector
```

```python
doc_result = embeddings.embed_documents([text, "This is another test sentence."])
doc_result[0]
```
